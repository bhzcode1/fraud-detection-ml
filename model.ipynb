```python
import numpy as np

def haversine_distance(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    earth_radius_km = 6371
    return earth_radius_km * c

new_unscaled_data['distance_km'] = haversine_distance(new_unscaled_data['lat'], new_unscaled_data['long'], new_unscaled_data['merch_lat'], new_unscaled_data['merch_long'])
```

```python
from sklearn.model_selection import train_test_split

X = new_unscaled_data2.drop('is_fraud', axis=1)
Y = new_unscaled_data2['is_fraud']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(Y_test.value_counts(normalize=True)*100)
print(Y_train.value_counts(normalize=True)*100)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
numeric_cols = ['amt', 'city_pop', 'age', 'distance_km', 'hours', 'month', 'day_of_week',
                 'merchant_freq', 'job_freq', 'city_freq', 'state_freq']


X_train[numeric_cols] = scalar.fit_transform(X_train[numeric_cols])


X_test[numeric_cols] = scalar.transform(X_test[numeric_cols])
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(class_weight= 'balanced', random_state=42)
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Classification Report:")
print(classification_report(Y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(Y_test, y_pred))
```

```python
from sklearn.metrics import average_precision_score
y_prob = model.predict_proba(X_test)[:, 1]
print(average_precision_score(Y_test, y_prob))
```
